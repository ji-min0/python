
## Mini project [행맨 게임 제작]

#### 과제 설명

이번 개발 주제는 '행맨' 게임이구요.

행맨 게임은 아주 유명한 단어 퀴즈 프로그램이예요

어떤 단어가 주어지면 그 단어의 길이만큼 빈 칸(밑줄) 이 주어지고
플레이어는 알파벳을 하나씩 얘기하면서 단어를 추측해요!

7번 틀리게 되면 게임이 끝납니다. 🥲



---

#### 예시를 들어볼께요!!

만약 E 가 단어에 포함돼있다면, 해당하는 위치에 E 를 표시해줍니다

_ _ _ _ E E
이렇게 되겠네요.

이번에는 A를 해볼께요
A 는 포함되지 않으므로 기회가 줄어들고 지금까지 맞힌 내용 그대로 출력합니다.

_ _ _ _ E E
변화가 없지요.

이번에는 C 를 해봅시다.

그럼
C _ _ _ E E

다음에는 F 를
C _ F F E E

마지막으로 O 를 해보면
C O F F E E 로 게임은 성공으로 끝나게 됩니다.

## 본격적인 행맨 게임 규칙 설명

- words 리스트에 추가된 단어들을 사용해주세요
  - words = ["apple", "banana", "orange", "grape", "lemon"]

- 단어의 길이에 맞게 밑줄을 출력해주세요  
  (예)banana의 경우 _ _ _ _ _ _

- 사용자부터 1글자씩 입력을 받되,
   - 단어에 입력값이 포함되면 "단어를 맞추셨군요. 더 힘내봐요"라고 출력해주세요"
   - 단어에 입력값이 포함되어 있지 않다면 "틀렸습니다. 남은 시도 횟수 "란 메시지와 함꼐 남은 횟수를 출력해주세요

- 매번 입력을 받을 때마다 현재까지 맞힌 글자들 표시(맞히면 글자는 밑줄 출력)
    - 예) a 입력 시 : a _ _ _ _
          p 입력 시 : a p p _ _
          c 입력 시 : a p p _ _

- 모든 단어를 맞췄을 경우 "축하합니다! 단어를 맞추셨습니다"와 시도 횟수를 출력해주세요


행맨은 정말 유명 게임으로 찾아볼 수 있는 자료들이 많습니다.   
문제가 어려우신 경우 구글링 통해 문제를 해결해보세요.  

```python
# 기본 라이브러리라 파이썬 설치시 자동으로 설치됩니다!
# random 모듈은 랜덤한 숫자를 생성할 때 사용합니다.
import random

# 게임에 사용될 단어 목록
words = ["apple", "banana", "orange", "grape", "lemon"]

# 행맨 그림
hangman_pics = [
    """
     ------
     |    |
     |
     |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |    |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|\
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|\
     |   /
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|\
     |   / \
     |
    ---""",
]

# 단어 랜덤 선택



class HangmanGame:
    def __init__(self):
        
        # 랜덤으로 단어를 선택하여 self.word에 저장
        self.word = random.choice(words)
        
        # 추측된 글자를 저장할 집합 self.guesses를 초기화
        self.guesses = set()
        
        # 남은 시도 횟수를 6으로 설정
        self.attemps = 6
        pass
    
    
    # 현재까지 맞춘 글자를 보여주는 함수
    def display(self):
        # 현재까지 맞춘 글자를 표시하는 문자열 result를 초기화
        result = ""
        
        # self.word의 각 글자를 순회
        for letter in self.word: 
            
            # 만약 그 글자가 self.guesses에 있으면 그대로 표시
            if letter in self.guesses: 
                result += letter + " "
            
            # 없으면 '_'로 표시
            else: 
                result += "_ "
        # 완성된 result 문자열을 반환
        return result.strip()
    
    
    # 게임 진행 함수
    def play(self):
        
        # 게임을 시작 (남은 시도 횟수 self.attempts가 0보다 클 동안 반복)
        while self.attemps > 0: 
            
            # 현재 행맨 그림 출력(틀린 횟수에 따라)
            wrong_guesses = 6 - self.attemps
            print(hangman_pics[wrong_guesses])
            
            # 현재 단어 상태 출력
            print(self.display)
            
            # 사용자에게 알파벳을 받음(대소문자 구분 안 함)
            guess = input("한 글자를 입력하세요: ").lower()
            
            # 입력받은 알파벳이 이미 추측된 경우 알림
            if guess in self.guesses: 
                print("이미 추측한 글자입니다. 다시 시도하세요.")
                continue
            
            # 새로운 추측이라면
            if guess in self.word: 
                
                # 단어에 포함되면 self.guesses에 추가하고
                self.guesses.add(guess)
                
                # "단어를 맞추셨군요. 더 힘내봐요!" 출력
                print("단어를 맞추셨군요. 더 힘내봐요!")
                
                # 단어의 모든 글자를 맞췄는지 확인하고
                if all(letter in self.guesses for letter in self.word): 
                    
                    # 맞췄으면 축하 메시지 후 종료
                    print("축하합니다! 단어를 맞추셨습니다.")
                    print(f"정답: {self.word}")
                    print(f"시도 횟수: {7 - self.attemps}")
                    break
            
            # 단어에 포함되지 않으면 시도 횟수를 1 감소시키고 틀렸다는 메시지 출력
            else: 
                self.attemps -= 1
                print(f"틀렸습니다. 남은 시도 횟수: {self.attemps}")
                
            # 반복문 종료 조건: 시도 횟수가 0
            if self.attemps == 0: 
                print(f"게임오버☠️ 정답은 '{self.word}' 였습니다.")
            
if __name__ == "__main__":
    game = HangmanGame()
    game.play()
```
