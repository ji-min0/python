### 📰 뉴스제목, 링크 가져오기 실습 🔗


```python
import requests
from bs4 import BeautifulSoup

# 다음에서 "죠스바" 뉴스 검색
url = 'https://search.daum.net/search?nil_suggest=btn&w=news&DA=SBC&cluster=y&q=%죠스바'

# 웹페이지 요청
response = requests.get(url)

# BeautifulSoup 객체 생성
soup = BeautifulSoup(response.content, 'lxml')

# CSS Selector로 뉴스 제목 찾기
title_tags = soup.select('.tit-g a')
print(f"총 {len(title_tags)}개 뉴스 제목 발견")

# 발견한 뉴스 제목, 링크 출력
for i, tag in enumerate(title_tags):

    # 제목에 있는 불필요한 공백 삭제
    title = tag.text.strip()

    #link url
    link = tag.get('href')

    print(f"\n{i + 1}. {title}")
    print(f"\n    URL: {link}")
    print("_"*60)
```
