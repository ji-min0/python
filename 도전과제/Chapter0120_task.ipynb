__[과제 1: 초기화 메서드 __init__]__

목표: __init__ 메서드를 구현하여 붕어빵 가게의 초기 재고, 가격, 판매량을 설정합니다.

설명: self.stock, self.prices, self.sales 딕셔너리를 초기화하세요.
- stock 
    - "팥붕어빵": 10
    - "슈크림붕어빵": 8
    - "초코붕어빵": 5

- prices
    - "팥붕어빵": 1000
    - "슈크림붕어빵": 1200
    - "초코붕어빵": 1500

- sales
    - "팥붕어빵": 0,
    - "슈크림붕어빵": 0,
    - "초코붕어빵": 0

```python
class BungeoppangShop:
    def __init__(self):
        
        # 재고 초기화
        self.stock = {
            "팥붕어빵": 10,
            "슈크림붕어빵": 8,
            "초코붕어빵": 5
        }
        
        # 가격 초기화
        self.prices = {
            "팥붕어빵": 1000,
            "슈크림붕어빵": 1200,
            "초코붕어빵": 1500
            }
        
        # 판매량 초기화
        self.sales = {
            "팥붕어빵": 0,
            "슈크림붕어빵": 0,
            "초코붕어빵": 0
        }
```

__[과제 2: 재고 확인 메서드 check_stock]__

목표: check_stock 함수를 구현하여 현재 붕어빵 재고를 출력합니다.

```python

    def check_stock(self): 
        print("남은 붕어빵: ")
        for flavor, count in self.stock.items():
            print(f"{flavor}: {count}개")
```

__[과제 3: 주문 처리 메서드 process_order]__

목표: process_order 함수를 구현하여 손님의 주문을 처리하고 재고를 업데이트합니다.

설명: bread_type과 bread_count를 매개변수로 받아, 재고를 확인하고 충분한 경우 재고를 감소시키고 판매량을 증가시킵니다. 재고가 부족하거나 잘못된 붕어빵 종류인 경우 적절한 메시지를 출력하세요.

```python
    def  process_order(self, bread_type, bread_count):
        try:
            # 재고 확인(재고가 주문량보다 많을 때만 주문 가능)
            if self.stock[bread_type] >= bread_count:
                
                # 재고 - 주문량
                self.stock[bread_type] -= bread_count
                
                # 판매량 + 주문량
                self.sales[bread_type] += bread_count
                
                # 판매 화면 출력
                print(f"{bread_type}: {bread_count}개를 판매했습니다.")
            
            # 재고 부족 화면 출력
            else:
                print(f"⛔️ {bread_type} 재고가 부족합니다")
        
        # 붕어빵 종류 이상하게 입력했을 때
        except KeyError:
            print("🚫 판매하지 않는 붕어빵입니다.")
```

__[과제 4: 관리자 모드 메서드 admin_mode]__

목표: admin_mode 메서드를 구현하여 관리자 모드에서 재고를 추가할 수 있도록 합니다.

설명: bread_type과 additional_stock을 입력받아 재고를 추가합니다. 잘못된 입력 값이 들어온 경우 적절한 메시지를 출력하세요.

```python
    def admin_mode(self):
        
        while True:
            
            # 붕어빵 종류 입력
            bread_type = input("추가할 붕어빵 종류를 입력하세요 (팥붕어빵, 슈크림붕어빵, 초코붕어빵)(종료 입력시 종료됨)")
            
            # 반복 탈출 단어 설정
            if bread_type == "종료":
                break
            
            try:
                # 추가할 수량 입력
                add_stock = int(input(f"추가할 {bread_type}의 개수를 입력하세요: "))
                
                # 재고 + 추가 수량
                self.stock[bread_type] += add_stock
                print(f"{bread_type} 재고가 {add_stock}개 추가되었습니다.")
                self.check_stock()
            
            # 입력 이상할 때
            except ValueError:
                print("🚫 잘못된 입력입니다. 숫자로 입력해주세요.")
            
            except KeyError:
                print("🚫 판매하지 않는 붕어빵입니다.")
```


__[과제 5: 매출 계산 메서드 calculate_total_sales]__

목표: calculate_total_sales 메서드를 구현하여 총 매출을 계산하고 출력합니다.

설명: self.sales와 self.prices를 사용하여 총 매출을 계산하고 출력하세요.

```python
def calculate_total_sales(self):
    
    # 총 매출 = 판매한 개수 * 가격
    total_sales = sum(self.sales[bread] * self.prices[bread] for bread in self.sales)
    
    print(f"총 매출: {total_sales}원")
    print("모든 거래 내역: ")
    
    # 빵마다 매출
    for bread in self.sales:
        bread_sales = self.sales[bread] * self.prices[bread]

        print(f"판매한 {bread}: {self.sales[bread]}개, 매출: {bread_sales}원")
```

__[과제 6: 메인 함수 main]__

목표: main 함수를 작성하여 프로그램의 주요 로직을 실행합니다.

설명:

```python
def main() :
    shop = BungeoppangShop()
    
    while True : 
        mode = input("원하는 모드를 선택하세요 (주문 ,관리자, 종료): ")
        
        if mode == "종료" :
            break
        
        if mode == "관리자" : 
            shop.admin_mode()
            continue
        
        if mode == "주문" : 
            
            while True:
                bread_type = input("주문할 붕어빵을 입력하세요 (팥붕어빵, 슈크림붕어빵, 초코붕어빵). (종료 입력 시 종료): ")
                
                if bread_type == "종료" :
                    break
                
                try :
                    bread_count = int(input("주문할 붕어빵 개수를 입력하세요: "))
                    shop.process_order(bread_type, bread_count)
                    
                except ValueError : 
                    print("🚫 잘못된 입력입니다. 숫자로 입력해주세요.")
                
                
                shop.check_stock()
    
    shop.calculate_total_sales()

if __name__ == "__main__" :
    main()
```
