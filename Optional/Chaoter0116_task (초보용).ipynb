__[문제 2: 붕어빵 주문 받기]__

손님이 주문한 붕어빵의 종류와 개수를 입력받는 기능을 구현하세요.

요구사항:
- 손님이 주문할 붕어빵의 종류와 개수를 입력받습니다.
- 주문 내용을 딕셔너리에 저장하세요.
- 주문 내역을 출력해주세요.

```python
order = {}

menu = {
    1 : "팥붕어빵",
    2 : "슈크림붕어빵",
    3 : "초코붕어빵"
}

# menu 출력
print("\n ===메뉴판===")
for num, flavor in menu.items():
    print(f'{num}. {flavor}')
print("0. 주문 끝내기")
    
while True :
    # menu 선택
    choice = int(input("\n번호를 선택하세요: "))
    
    # 주문 끝내기
    if choice == 0 :
        break
    
    # 이상한 번호 입력 시
    if choice not in menu :
        print("잘못된 선택입니다. 다시 입력해주세요")
        continue
    
    # 주문 개수 입력
    fish_flavor = menu[choice]
    fish_count = int(input(f"{fish_flavor} 몇 개 주문하시겠어요?"))
    
    # 주문 내용 저장
    order[fish_flavor] = order.get(fish_flavor, 0) + fish_count
    
# 주문 내역 출력
print("\n=========")
for fish_flavor, fish_count in order.items() :
    print(f"주문내역: {fish_flavor} {fish_count}개")
```

__[문제 3: 붕어빵 주문 처리]__

손님의 주문을 처리하여 붕어빵 재고를 업데이트하는 기능을 구현하세요.

요구사항:
- 손님의 주문 내용을 기반으로 재고를 업데이트합니다.
- 재고가 부족할 경우, 손님에게 재고 부족을 알리고 재고를 감소시키지 않습니다.
- 재고가 충분할 경우, 주문한 만큼 재고를 감소시키고 판매를 완료합니다.
- 판매가 완료된 경우 판매된 붕어빵 맛과 개수를 출력하세요.

```python
if fish_count > stock[fish_flavor]:
    print(f"[재고 부족] {fish_flavor}는 현재 주문하신 수량만큼 없습니다.")
else:
    stock[fish_flavor] -= fish_count
    print(f"[판매 완료] {fish_flavor} {fish_count}개 판매되었습니다!")
```

__[문제 4: 여러 손님 주문 처리]__

여러 손님들의 주문을 처리할 수 있도록 시스템을 확장하세요.

요구사항:
- 손님이 "종료"를 입력할 때까지 주문을 계속 받습니다.
- 각 주문마다 재고를 업데이트합니다.
- 현재 재고를 화면에 출력합니다.

```python
# 재고
stock = {
    "팥붕어빵": 10,
    "슈크림붕어빵": 8,
    "초코붕어빵": 5
}

# 붕어빵 종류
menu = {
    1: "팥붕어빵",
    2: "슈크림붕어빵",
    3: "초코붕어빵"
}

# 메뉴 출력
print("=== 붕어빵 메뉴 ===")
for num, flavor in menu.items():
    print(f"{num}. {flavor}")
print("0. 종료")


while True:
    # 붕어빵 맛 입력
    choice = input("\n번호를 선택하세요 (0: 종료): ")
    
    # 0 (종료 입력시)
    if choice == "0" or choice.lower() == "종료":
        print("\n주문을 종료합니다.")
        break
    
    # 이상하게 입력시
    if not choice.isdigit() or int(choice) not in menu:
        print("\n잘못된 선택입니다. 다시 입력해주세요.")
        continue
    
    choice = int(choice)
    fish_flavor = menu[choice]
    
    # 주문 수량 입력
    fish_count = input(f"{fish_flavor} 몇 개 주문하시겠어요? ")
    if not fish_count.isdigit():
        print("\n수량은 숫자로 입력해주세요.")
        continue
    fish_count = int(fish_count)
    
    # 재고 확인, 처리
    if fish_count > stock[fish_flavor]:
        print(f"[재고 부족] {fish_flavor}는 현재 주문하신 수량만큼 없습니다.")
    
    # 판매 가능
    else:
        stock[fish_flavor] -= fish_count
        print(f"[판매 완료] {fish_flavor} {fish_count}개 판매되었습니다!")
        
    # 현재 재고
    print("\n현재 남은 수량")
    for flavor, count in stock.items() : 
        print(f'{flavor}: {count}개')
```

__[문제 5: 관리자 모드]__

관리자 모드를 통해 붕어빵 재고를 추가할 수 있는 기능을 구현하세요.

요구사항:
- 관리자는 특정 붕어빵의 재고를 추가할 수 있습니다.
- "종료"를 입력하면 관리자 모드를 종료합니다.
- 추가된 재고를 업데이트하고 현재 재고를 화면에 출력합니다.

```python
# 재고
stock = {
    "팥붕어빵": 10,
    "슈크림붕어빵": 8,
    "초코붕어빵": 5
}

# 붕어빵 종류
menu = {
    1: "팥붕어빵",
    2: "슈크림붕어빵",
    3: "초코붕어빵"
}

print("====(관리자모드)====")
print("\n재고를 추가하려는 붕어빵 번호를 선택하세요")
print("\n=== 붕어빵 메뉴 ===")
for num, flavor in menu.items():
    print(f"{num}. {flavor}")
print("\n'종료' 입력 시 관리자 모드를 종료합니다")

while True:
    # 추가할 맛 선택
    fish_input = input("\n재고를 추가하려는 붕어빵 번호를 선택하세요 (종료 입력 시 종료): ")

    # 종료 처리
    if fish_input == "종료":
        print("\n관리자 모드를 종료합니다.")
        break

    # 입력 번호 → 붕어빵 이름 변환
    if fish_input.isdigit() and int(fish_input) in menu:
        fish_flavor = menu[int(fish_input)]
    else:
        print("\n잘못된 선택입니다. 다시 입력해주세요.")
        continue

    # 추가할 수량 입력
    add_count = input(f"\n{fish_flavor}의 추가할 개수를 입력하세요: ")
    if not add_count.isdigit():
        print("\n재고 수량은 숫자로 입력해주세요.")
        continue
    add_count = int(add_count)

    # 재고 업데이트
    stock[fish_flavor] += add_count
    print(f"\n{fish_flavor}의 재고가 {add_count}개 추가되었습니다.")

    # 현재 재고 출력
    print("\n현재 남은 수량:")
    for flavor, quantity in stock.items():
        print(f"{flavor}: {quantity}개")
```

__[문제 6: 매출 확인]__

붕어빵 판매 가격을 설정하고, 판매한 붕어빵의 매출을 계산하세요.

요구사항:
- 각 붕어빵의 가격을 딕셔너리로 관리합니다.
- 판매된 붕어빵의 종류와 개수를 딕셔너리로 관리합니다.
- 총 매출을 계산하여 출력합니다.

초기 설정:
- 가격
    - "팥붕어빵": 1000
    - "슈크림붕어빵": 1200
    - "초코붕어빵": 1500

- 판매 개수
    - "팥붕어빵": 0
    - "슈크림붕어빵": 0
    - "초코붕어빵": 0

```python
prices = {
    "팥붕어빵": 1000,
    "슈크림붕어빵": 1200,
    "초코붕어빵": 1500
}
sales = {
    "팥붕어빵": 0,
    "슈크림붕어빵": 0,
    "초코붕어빵": 0
}

total_sales = sum(sales[flavor] * prices[flavor] for flavor in sales)
print(f"총 매출: {total_sales}원")
```

```python
# 최종 코드


# 재고
stock = {
    "팥붕어빵": 10,
    "슈크림붕어빵": 8,
    "초코붕어빵": 5
}

# 붕어빵 종류
menu = {
    1: "팥붕어빵",
    2: "슈크림붕어빵",
    3: "초코붕어빵"
}

# 가격
prices = {
    "팥붕어빵": 1000,
    "슈크림붕어빵": 1200,
    "초코붕어빵": 1500
}

# 주문 내역 & 매출
order = {}
sales = {
    "팥붕어빵": 0,
    "슈크림붕어빵": 0,
    "초코붕어빵": 0
}






# 관리자 모드
print("====(관리자모드)====")
while True:
    print("\n=== 붕어빵 메뉴 ===")
    for num, flavor in menu.items():
        print(f"{num}. {flavor}")
    print("\n'종료' 입력 시 관리자 모드를 종료합니다")

    fish_input = input("\n재고를 추가하려는 붕어빵 번호를 선택하세요 (종료 입력 시 종료): ")

    if fish_input == "종료":
        print("\n관리자 모드를 종료합니다.\n")
        break

    if fish_input.isdigit() and int(fish_input) in menu:
        fish_flavor = menu[int(fish_input)]
    else:
        print("\n잘못된 선택입니다. 다시 입력해주세요.")
        continue

    add_count = input(f"\n{fish_flavor}의 추가할 개수를 입력하세요: ")
    if not add_count.isdigit():
        print("\n재고 수량은 숫자로 입력해주세요.")
        continue
    add_count = int(add_count)

    stock[fish_flavor] += add_count
    print(f"\n{fish_flavor}의 재고가 {add_count}개 추가되었습니다.")

    print("\n현재 남은 수량:")
    for flavor, quantity in stock.items():
        print(f"{flavor}: {quantity}개")






# 주문 모드
print("==== 주문 모드 ====")
while True:
    print("\n=== 메뉴 ===")
    for num, flavor in menu.items():
        print(f"{num}. {flavor}")
    print("0. 주문 끝내기")

    fish_input = input("\n번호를 선택하세요: ")

    if fish_input == "0":
        print("\n주문을 종료합니다.")
        break

    if not fish_input.isdigit() or int(fish_input) not in menu:
        print("\n잘못된 선택입니다. 다시 입력해주세요.")
        continue

    fish_flavor = menu[int(fish_input)]

    fish_count = input(f"{fish_flavor} 몇 개 주문하시겠어요? ")
    if not fish_count.isdigit():
        print("\n수량은 숫자로 입력해주세요.")
        continue
    fish_count = int(fish_count)

    if fish_count > stock[fish_flavor]:
        print(f"[재고 부족] {fish_flavor}는 현재 주문하신 수량만큼 없습니다.")
    else:
        stock[fish_flavor] -= fish_count
        order[fish_flavor] = order.get(fish_flavor, 0) + fish_count
        sales[fish_flavor] += fish_count
        print(f"[판매 완료] {fish_flavor} {fish_count}개 판매되었습니다!")

    print("\n현재 남은 수량:")
    for flavor, quantity in stock.items():
        print(f"{flavor}: {quantity}개")






# 최종 주문 내역 & 매출
print("\n=========")
for fish_flavor, fish_count in order.items():
    print(f"주문내역: {fish_flavor} {fish_count}개")

total_sales = sum(sales[fish_flavor] * prices[fish_flavor] for fish_flavor in sales)
print(f"총 매출: {total_sales}원")
```
